```python
seasons = [str(yr) + "-" + str(yr+1)[2:4] for yr in range(1990, 2023)]
teams = pd.read_csv(db_path + "teams.csv").id.to_list()
types = ["Regular Season","Pre Season","Playoffs","Preseason"]
list = []
for season in seasons:
    print (season)
    for team in teams:
        for type in types:
            for game in teamgamelog.TeamGameLog(season=season,season_type_all_star=type,team_id=team).get_dict()["resultSets"][0]["rowSet"]:
                list.append({"game_id":game[1],"season":season,"game_type":type,"team_id":game[0],"game_date":game[2],"matchup":game[3],"WL":game[4]})
```

```python
games = pd.DataFrame(list)
games
```

```python
len(games["game_id"].unique())
```

```python
maybe = games.drop_duplicates(ignore_index=True)
maybe
```

```python
len(maybe["game_id"].unique())
```

```python
for game in maybe["game_id"].unique():
    subset = maybe.loc[maybe["game_id"]==game,]
    if (len(subset.index) != 2):
        print (subset)
```

```python
temp = []
for game in maybe["game_id"].unique():
    subset = maybe.loc[maybe["game_id"]==game,].reset_index()
    if (len(subset.index) != 2):
        continue
    if ("vs." in subset.at[0,"matchup"]):
        temp.append({"game_id":subset.at[0,"game_id"],"season":subset.at[0,"season"],"game_date":subset.at[0,"game_date"],
                    "h_team_id":subset.at[0,"team_id"],"a_team_id":subset.at[1,"team_id"],"matchup":subset.at[0,"matchup"]})
    else:
        temp.append({"game_id":subset.at[0,"game_id"],"season":subset.at[0,"season"],"game_date":subset.at[0,"game_date"],
                    "h_team_id":subset.at[1,"team_id"],"a_team_id":subset.at[0,"team_id"],"matchup":subset.at[1,"matchup"]})
final = pd.DataFrame(temp)
final
```

```python
final["game_date"] = pd.to_datetime(final["game_date"])
final = final.sort_values(by="game_date")
```

```python
final.to_csv(db_path + "games.csv", index = False)
```

```python
from nba_api.stats.endpoints import boxscoreadvancedv3

def advanced_boxscore_table():
    games = pd.read_csv(db_path + "games.csv")
    #valid box score years
    seasons = [str(yr) + "-" + str(yr+1)[2:4] for yr in range(1996, 2023)]
    string_s = seasons[0]
    for season in seasons:
        string_s += "|" + season

    player_table = []
    team_table = []
    game_ids = games[games["season"].str.contains(string_s[8:])]["game_id"]
    for game in tqdm(game_ids[0:10]):
        #time.sleep(0.5)
        try:
            a = boxscoreadvancedv3.BoxScoreAdvancedV3(game_id="00"+str(game)).get_dict()
        except AttributeError:
            print ("Error: 00" + str(game))
            continue
        for side in ["homeTeam","awayTeam"]:
            temp_t = {"game_id":a['boxScoreAdvanced']['gameId'],"team_id":a['boxScoreAdvanced'][side]["teamId"]}
            for key in a['boxScoreAdvanced'][side]['statistics']:
                temp_t[key] = a['boxScoreAdvanced'][side]['statistics'][key]
            team_table.append(temp_t)
            for player in a['boxScoreAdvanced'][side]["players"]:
                temp_p = {"game_id":a['boxScoreAdvanced']['gameId'],"player_id":player['personId'],"team_id":a['boxScoreAdvanced'][side]["teamId"]}
                for key in player['statistics']:
                    temp_p[key] = player['statistics'][key]
                player_table.append(temp_p)
    
    df = pd.DataFrame(team_table)
    df.to_csv(db_path+"advanced_boxscores_teams.csv",index=False)

    df = pd.DataFrame(player_table)
    df.to_csv(db_path+"advanced_boxscores_players.csv",index=False)
```

```python
advanced_boxscore_table()
```
